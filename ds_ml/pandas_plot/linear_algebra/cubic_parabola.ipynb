```python
import matplotlib.pyplot as plt


def p(x):
    y = x ** 3 - x + 2
    return y


x = []
for number in range(-5, 6):
    x.append(number)
print(x)
y = []
for value in x:
    y.append(p(value))
print(y)

plt.plot(x)
plt.xlabel("x_p--->")
plt.show()

plt.plot(y)
plt.ylabel("y_value--->")
plt.show()

plt.plot(x, y)
plt.xlabel("x--->")
plt.ylabel("y--->")
plt.title("cubic_polynomial")
plt.show()
```
