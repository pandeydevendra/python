```python
#y=x+2
#y=-x+2

import matplotlib.pyplot as plt

def get_y1(x):
    return (2*x-2)

def get_y2(x):
    return (-x+2)

x = []
for i in range(-2,3):
    x.append(i)
print(x)   
    
y1=[]
y2=[]

for px in x:
    py1 = get_y1(px)
    py2 = get_y2(px)
    y1.append(py1)
    y2.append(py2)
    
    

plt.plot(x, y1, color='green', linestyle='dashed', linewidth = 2, 
         marker='o', markerfacecolor='blue', markersize=10, label="y1")     

plt.plot(x, y2, color='red', linestyle='dashed', linewidth = 2, 
         marker='.', markerfacecolor='black', markersize=10, label="y2") 


plt.xlabel('x - axis') 

plt.ylabel('y - axis') 
 
plt.legend()
plt.show()
```
