```python
import matplotlib.pyplot as plt

def get_h(t):
    h=(1/2)*9.81*(t**2)
    return h


t_list = []
for i in range (20):
    t_list.append(i)
h_list = []

for t in t_list:
    h=get_h(t)
    h_list.append(h)
    
plt.plot(t_list,h_list, color='green', linestyle='dashed', linewidth = 2, 
         marker='o', markerfacecolor='blue', markersize=10)     

    
plt.xlabel('t - axis') 

plt.ylabel('h - axis') 
plt.show() 
```
