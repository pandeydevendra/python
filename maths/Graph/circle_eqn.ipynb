```python
import matplotlib.pyplot as plt
import math
import numpy as np 

r=2
def get_y1(x):
    y1 = math.sqrt((r**2) -(x**2))
    return y1

def get_y2(x):
    y2 = -math.sqrt((r**2) -(x**2))
    return y2

x = np.arange(-r, r, 0.1)   
    
y1=[]
y2=[]
for px in x:
    py1 = get_y1(px)
    y1.append(py1)
    py2 = get_y2(px)
    y2.append(py2)

plt.plot(x, y1, color='green', linestyle='dashed', linewidth = 2, 
         marker='.', markerfacecolor='blue', markersize=4, label="x**2 + y**2 = r**2")     
plt.plot(x, y2, color='red', linestyle='dashed', linewidth = 2, 
         marker='.', markerfacecolor='blue', markersize=4, label="x**2 + y**2 = r**2")  

plt.xlabel('x - axis') 

plt.ylabel('y - axis') 
 
plt.legend()
plt.show()
```
